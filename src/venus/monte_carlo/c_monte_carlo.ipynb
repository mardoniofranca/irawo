```python
import random
import math
import matplotlib.pyplot as plt
```

```python
# Número de pontos a serem gerados
n = 1000000

# Raio do círculo
r = 1.0

# Contadores
num_dentro = 0
num_fora = 0

# Listas para armazenar coordenadas x e y dos pontos gerados
x_dentro = []
y_dentro = []
x_fora = []
y_fora = []
```

```python
# Gerar pontos aleatórios
for i in range(n):
    x = random.uniform(-r, r)
    y = random.uniform(-r, r)

    # Verificar se o ponto está dentro do círculo
    if math.sqrt(x**2 + y**2) <= r:
        num_dentro += 1
        x_dentro.append(x)
        y_dentro.append(y)
    else:
        num_fora += 1
        x_fora.append(x)
        y_fora.append(y)
```

```python
# Calcular o valor de pi
pi = 4 * num_dentro / n

# Imprimir resultado
print("Valor de pi:", pi)
```

```python
# Plotar gráfico
fig, ax = plt.subplots()
ax.set_aspect('equal', adjustable='box')
ax.scatter(x_dentro, y_dentro, c='blue', s=1)
ax.scatter(x_fora, y_fora, c='red', s=1)
ax.set_title('Simulação de Monte Carlo para cálculo de pi')
ax.set_xlabel('Coordenada x')
ax.set_ylabel('Coordenada y')
ax.set_xlim(-r, r)
ax.set_ylim(-r, r)
plt.show()
```
