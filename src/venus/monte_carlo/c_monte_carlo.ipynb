## Simulação Monte Carlo para o Cálculo do número Pi

```python
import random
import math
import matplotlib.pyplot as plt
```

##### 1. Primeiro, são importadas as bibliotecas necessárias para o cálculo de pi e para a apresentação da visualização (random, math e matplotlib.pyplot).

##### 2. Em seguida, são definidas as variáveis n (número de pontos a serem gerados) e r (raio do círculo).

##### 3. São criados os contadores num_dentro e num_fora para contar o número de pontos gerados dentro e fora do círculo, respectivamente, e as listas x_dentro, y_dentro, x_fora e y_fora para armazenar as coordenadas x e y dos pontos gerados.

##### 4. É feita uma iteração de n vezes para gerar pontos aleatórios no plano cartesiano com coordenadas x e y variando de -r a r.

##### 5. Em cada iteração, é verificado se o ponto gerado está dentro ou fora do círculo. Se o ponto estiver dentro do círculo, o contador num_dentro é incrementado e as coordenadas x e y do ponto são adicionadas às listas x_dentro e y_dentro, respectivamente. Caso contrário, o contador num_fora é incrementado e as coordenadas x e y do ponto são adicionadas às listas x_fora e y_fora, respectivamente.

##### 6. Ao final da iteração, é calculado o valor de pi utilizando a fórmula pi = 4 * num_dentro / n.

##### 7. O valor de pi é impresso na tela.

##### 8. É plotado um gráfico com os pontos gerados, em que os pontos dentro

```python
# Número de pontos a serem gerados
n = 1000000

# Raio do círculo
r = 1.0

# Contadores
num_dentro = 0
num_fora = 0

# Listas para armazenar coordenadas x e y dos pontos gerados
x_dentro = []
y_dentro = []
x_fora = []
y_fora = []
```

```python
# Gerar pontos aleatórios
for i in range(n):
    x = random.uniform(-r, r)
    y = random.uniform(-r, r)

    # Verificar se o ponto está dentro do círculo
    if math.sqrt(x**2 + y**2) <= r:
        num_dentro += 1
        x_dentro.append(x)
        y_dentro.append(y)
    else:
        num_fora += 1
        x_fora.append(x)
        y_fora.append(y)
```

```python
# Calcular o valor de pi
pi = 4 * num_dentro / n

# Imprimir resultado
print("Valor de pi:", pi)
```

```python
# Plotar gráfico
fig, ax = plt.subplots()
ax.set_aspect('equal', adjustable='box')
ax.scatter(x_dentro, y_dentro, c='blue', s=1)
ax.scatter(x_fora, y_fora, c='red', s=1)
ax.set_title('Simulação de Monte Carlo para cálculo de pi')
ax.set_xlabel('Coordenada x')
ax.set_ylabel('Coordenada y')
ax.set_xlim(-r, r)
ax.set_ylim(-r, r)
plt.show()
```
