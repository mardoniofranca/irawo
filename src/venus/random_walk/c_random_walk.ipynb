```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set the number of steps and the number of simulations
n_steps = 10000
n_simulations = 10
```

```python
# generate random steps for each simulation
steps = 2 * np.random.randint(0, 2, (n_simulations, n_steps)) - 1
```

```python
# calculate the position of each simulation at each step
positions = np.cumsum(steps, axis=1)
```

```python
# plot the results
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title("Random Walks Simulation")
ax.set_xlabel("Steps")
ax.set_ylabel("Position")
for i in range(n_simulations):
    ax.plot(positions[i], label=f"Simulation {i+1}")
ax.legend()
plt.show()
```

## n_simulations random walks with n_steps steps each, calculates the standard deviation and variance of the final positions

This code generates n_simulations random walks with n_steps steps each, calculates the standard deviation and variance of the final positions, and plots the resulting random walks and histogram of the final positions.

The std variable is the standard deviation of the final positions across all simulations, and var is the variance of the final positions across all simulations. These values are displayed as text annotations on the plot.

The histogram of the final positions is generated using the plt.hist() function, which bins the final positions into 20 equally spaced bins and plots the counts of the final positions in each bin.

You can adjust the values of n_steps and n_simulations to generate more or fewer steps and simulations. Additionally, you can customize the plot by adjusting the labels, title, and other parameters of the ax object and the plt.hist() function.

```python
# set the number of steps and the number of simulations
n_steps = 1000
n_simulations = 100

# generate random steps for each simulation
steps = 2 * np.random.randint(0, 2, (n_simulations, n_steps)) - 1

# calculate the position of each simulation at each step
positions = np.cumsum(steps, axis=1)

# calculate the standard deviation and variance of the final positions
std = np.std(positions[:, -1])
var = np.var(positions[:, -1])
```

```python
# plot the results
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title("Random Walks Simulation")
ax.set_xlabel("Steps")
ax.set_ylabel("Position")
for i in range(n_simulations):
    ax.plot(positions[i], alpha=0.5)
ax.text(0.05, 0.9, f"Final position mean: {positions[:, -1].mean():.2f}", 
        transform=ax.transAxes, fontsize=12)
ax.text(0.05, 0.85, f"Final position std: {std:.2f}", 
        transform=ax.transAxes, fontsize=12)
ax.text(0.05, 0.8, f"Final position variance: {var:.2f}", 
        transform=ax.transAxes, fontsize=12)
plt.hist(positions[:, -1], bins=20, alpha=0.5)
plt.show()
```
